```python
# b

x = np.linspace(-10, 10, num=100)
y1 = np.array([-1/j for j in x])
y2 = np.array([abs(j) for j in x])
y3 = np.array([j ** 3 for j in x])

plt.plot(x, y1)
plt.xlabel('x')
plt.ylabel('y1')
plt.text(-10, 7.5, "Graph Type: y = 1/x")
plt.show()

plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y2')
plt.text(-2.5, 9, "Graph Type: y = |x|")
plt.show()

plt.plot(x, y3)
plt.xlabel('x')
plt.ylabel('y3')
plt.text(4, -1000,"Graph Type: y = x^3")
plt.show()
```

```python
# (c)

x = np.linspace(-10, 10, num=100)
y1 = np.array([-j ** 2 for j in x])
y2 = np.array([j ** 2 for j in x])
y3 = np.array([abs(j) for j in x])

fig, (graph1, graph2, graph3) = plt.subplots(3)
graph1.plot(x, y1)
graph2.plot(x, y2)
graph3.plot(x, y3)
# Review what this does
# plt.subplots_adjust(top=2, bottom=1, left=1, right=3, hspace=0.25, wspace=0.35)
plt.subplots_adjust(hspace=0.5)
plt.show()

# adjusted subplots
fig, (graph1, graph2, graph3) = plt.subplots(1, 3, figsize=(10, 5))

tmrFont = {"fontname": "Times New Roman"}
fig.suptitle("Horizontally Stacked Subplots", **tmrFont, fontsize = 20)

graph1.plot(x, y1)
graph2.plot(x, y2)
graph3.plot(x, y3)

plt.show()
```
